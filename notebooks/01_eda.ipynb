```python
# ==============================
# 📊 COMPLETE EDA PIPELINE
# ==============================

# 📦 Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

# ⚙️ Config
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

RANDOM_STATE = 42
SAMPLE_SIZE = 10000

# 📥 Load Data
df = pd.read_csv("../../data/student_mental_health_dataset.csv")
data = df.sample(SAMPLE_SIZE, random_state=RANDOM_STATE)

print("Shape:", data.shape)
display(data.head())

# 🔍 Basic Inspection
print("\n--- INFO ---")
data.info()

print("\n--- DESCRIPTION ---")
display(data.describe())

print("\n--- MISSING VALUES ---")
print(data.isnull().sum())

print("\n--- DUPLICATES ---")
print(data.duplicated().sum())

# 📊 Column Types
cat_cols = data.select_dtypes(include="object").columns
num_cols = data.select_dtypes(include="number").columns

print("\nCategorical Columns:", list(cat_cols))
print("Numerical Columns:", list(num_cols))

# ==============================
# 📈 Numerical Features
# ==============================
n_cols = 3
n_rows = math.ceil(len(num_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(data[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f"{col} Distribution", fontweight="bold")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# ==============================
# 📊 Categorical Features
# ==============================
n_cols = 3
n_rows = math.ceil(len(cat_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.countplot(x=data[col], ax=axes[i], alpha=0.7)
    axes[i].tick_params(axis="x", rotation=45)
    axes[i].set_title(f"{col} Distribution", fontweight="bold")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# ==============================
# 🔗 Relationship Analysis
# ==============================
n_cols = 2
n_rows = math.ceil(len(cat_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 6, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    sns.boxplot(x=data[col], y=data["dropout_risk"], ax=axes[i])
    axes[i].set_title(f"{col} vs Dropout Risk", fontweight="bold")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# ==============================
# ⚠️ Outlier Detection
# ==============================
n_cols = 3
n_rows = math.ceil(len(num_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.boxplot(y=data[col], ax=axes[i])
    axes[i].set_title(col, fontweight="bold")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# ==============================
# 🔥 Correlation Heatmap
# ==============================
corr = data[num_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap", fontsize=18, fontweight="bold")
plt.tight_layout()
plt.show()

# ==============================
# 🎯 Target Creation
# ==============================
data["dropout_labels"] = pd.qcut(
    data["dropout_risk"],
    q=3,
    labels=["Low", "Medium", "High"]
)

data.drop("dropout_risk", axis=1, inplace=True)

print("\nTarget Distribution:")
print(data["dropout_labels"].value_counts())
```
